```python
from datetime import datetime
import ast
import pandas as pd

movies_df = pd.read_csv('tmdb_5000_movies.csv')

def extract_genre_names(genres_str):
    try:
        genres_list = ast.literal_eval(genres_str)
        genre_names = [genre['name'] for genre in genres_list]
        return ', '.join(genre_names)
    except (ValueError, SyntaxError):
        return ''

movies_df['genre_names'] = movies_df['genres'].apply(extract_genre_names)


def get_time_of_day():
    current_hour = datetime.now().hour
    if 6 <= current_hour < 12:
        return 'morning'
    elif 12 <= current_hour < 17:
        return 'afternoon'
    elif 17 <= current_hour < 21:
        return 'evening'
    elif 21 <= current_hour < 24:
        return 'night'
    else:
        return 'late_night'

def adjust_recommendations_by_time(recommendations, time_segment):
    if time_segment == 'morning':
        recommendations = recommendations[recommendations['genre_names'].str.contains('Comedy|Drama', na=False)]
    elif time_segment == 'night':
        recommendations = recommendations[recommendations['genre_names'].str.contains('Thriller|Drama', na=False)]
    return recommendations

recommendations = movies_df[['title', 'genre_names']]

time_segment = get_time_of_day()
recommendations_adjusted = adjust_recommendations_by_time(recommendations, time_segment)

print(recommendations_adjusted.head())
```

```python
print(recommendations.columns)

def adjust_recommendations_by_emotion(recommendations, detected_emotion):
    """Adjust movie recommendations based on detected emotion."""
    if detected_emotion == 'joy':
        return recommendations[recommendations['genre_names'].str.contains('Comedy|Feel-Good', case=False)]
    elif detected_emotion == 'sadness':
        return recommendations[recommendations['genre_names'].str.contains('Drama|Inspirational', case=False)]
    else:
        return recommendations

recommendations_adjusted = adjust_recommendations_by_emotion(recommendations, detected_emotion)

print(recommendations_adjusted)
```

```python
import requests

def get_weather_data(api_key, location='London'):
    """Fetch current weather data from OpenWeatherMap."""
    base_url = "http://api.openweathermap.org/data/2.5/weather"

    params = {
        'q': location,
        'appid': api_key,
        'units': 'metric'
    }

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        weather_data = response.json()
        main_weather = weather_data['weather'][0]['main']
        description = weather_data['weather'][0]['description']
        temperature = weather_data['main']['temp']
        return {
            'weather': main_weather,
            'description': description,
            'temperature': temperature
        }
    else:
        print(f"Failed to fetch weather data: {response.status_code}")
        return None

api_key = 'c44a28e492fae618134775ef6329082a'
weather_info = get_weather_data(api_key, location='New York')
if weather_info:
    print(f"Weather: {weather_info['weather']}")
    print(f"Description: {weather_info['description']}")
    print(f"Temperature: {weather_info['temperature']}°C")
```

##Number of Movie Recommendations by Genre Based on Current Weather

```python


import requests
import matplotlib.pyplot as plt

def fetch_real_time_weather(api_key, city):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        weather_info = {
            'description': data['weather'][0]['description'],
            'temperature': data['main']['temp']
        }
        return weather_info
    else:
        print(f"Error: Unable to fetch weather data (Status code: {response.status_code})")
        return None

def group_small_categories(genre_counts, threshold=3):
    grouped_genre_counts = genre_counts.copy()
    small_categories = grouped_genre_counts[grouped_genre_counts < threshold]
    grouped_genre_counts = grouped_genre_counts[grouped_genre_counts >= threshold]
    grouped_genre_counts['Others'] = small_categories.sum()
    return grouped_genre_counts

def adjust_recommendations_by_weather(recommendations, weather_info):
    return recommendations

def plot_weather_based_recommendations(recommendations, weather_info, threshold=3):
    weather_adjusted = adjust_recommendations_by_weather(recommendations, weather_info)
    genre_counts = weather_adjusted['genre_names'].str.split(',').explode().value_counts()
    grouped_genre_counts = group_small_categories(genre_counts, threshold=threshold)
    plt.figure(figsize=(10, 10))
    pie_chart = grouped_genre_counts.plot.pie(
        autopct=lambda p: f'{p:.1f}%' if p > threshold else '',
        textprops={'fontsize': 12},
        startangle=90,
        colormap='cool',
        labels=None
    )

    plt.legend(
        grouped_genre_counts.index,
        title="Genres",
        loc="center left",
        bbox_to_anchor=(1, 0.5)
    )

    plt.title(f"Recommended Genres Under {weather_info['description']} ({weather_info['temperature']}°C)", fontsize=14)
    plt.ylabel("")
    plt.tight_layout()
    plt.show()

api_key = '0db371c9dc41ee8b49cb7d22f6cce9e0'
city = 'Gurugram'
weather_info = fetch_real_time_weather(api_key, city)

if weather_info:
    plot_weather_based_recommendations(movies_df[['title', 'genre_names']], weather_info, threshold=3)
```

##Impact of Emotion and Weather on Movie Genre Recommendations



```python
def plot_emotion_weather_impact(recommendations, emotions, weather_conditions):
    data = {}
    for emotion in emotions:
        emotion_recs = adjust_recommendations_by_emotion(recommendations, emotion)
        for weather in weather_conditions:
            weather_recs = adjust_recommendations_by_weather(emotion_recs, weather)
            genre_counts = weather_recs['genre_names'].str.split(',').explode().value_counts()
            data[(emotion, weather['description'])] = genre_counts

    combined_df = pd.DataFrame(data).fillna(0)
    combined_df.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='tab10')
    plt.title("Impact of Emotion and Weather on Genre Recommendations")
    plt.xlabel("Genre")
    plt.ylabel("Frequency")
    plt.legend(title="Emotion & Weather", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


emotions = ['joy', 'anger', 'sadness']
weather_conditions = [{'description': 'rain'}, {'description': 'clear skies'}, {'description': 'snow'}]
plot_emotion_weather_impact(movies_df[['title', 'genre_names']], emotions, weather_conditions)
```

##Compare how different weather conditions affect the popularity of movie genres.


```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_weather_genre_bar_chart(recommendations, weather_conditions):
    genre_counts_by_weather = {}

    for weather in weather_conditions:
        weather_adjusted = adjust_recommendations_by_weather(recommendations, {'description': weather})
        genre_counts = weather_adjusted['genre_names'].str.split(',').explode().value_counts()
        genre_counts_by_weather[weather] = genre_counts

    genre_df = pd.DataFrame(genre_counts_by_weather).fillna(0)

    genre_df.plot(kind='bar', figsize=(12, 6), colormap='coolwarm', width=0.8)
    plt.title("Genre Popularity Across Weather Conditions")
    plt.xlabel("Genre")
    plt.ylabel("Count of Recommendations")
    plt.xticks(rotation=45)
    plt.legend(title="Weather Conditions")
    plt.tight_layout()
    plt.show()

weather_conditions = ['clear', 'rainy', 'cloudy']
plot_weather_genre_bar_chart(movies_df[['title', 'genre_names']], weather_conditions)
```
